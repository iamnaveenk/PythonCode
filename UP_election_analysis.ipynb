```python
df=pd.concat([df1, df2], axis=1)
df
```

```python
df.info()
```

```python
#How many constituencies did the Bharatiya Janata Party (BJP) win?

df[(df['Rank']==1) & (df['Party']=='Bharatiya Janata Party')].value_counts().sum()
```

```python
#Similarly calculate the number of constituencies won by each party.
df[(df['Rank']==1)]['Party'].value_counts()
```

```python
#In how many constituencies did the Bharatiya Janata Party and the Samajwadi Party occupy the first two ranks (i.e either BJP comes first and SP second or vice versa)

df[(df['Rank']==1 | df['Rank']==2) & (df['Party']=='Bharatiya Janata Party' | df['Party']=='Samajwadi Party')].value_counts().sum()
```
